```python
import pandas as pd
import numpy as np

# read in compustat/crsp linking table
link=pd.read_csv('quant_analysis/a6f84adbf39c2211.csv')
link =link.rename(columns={"LPERMNO": "PERMNO"})
link['fyear'] = pd.to_datetime(link['fyear'], format='%Y').dt.to_period('Y')
link.head()
```

```python
# read in compustat data
compustat_df = pd.read_csv('quant_analysis/72b0e3ef6d338179.csv')
compustat_df = compustat_df.rename(columns={"gvkey": "GVKEY"})
compustat_df['fyear'] = pd.to_datetime(compustat_df['fyear'], format='%Y').dt.to_period('Y')
compustat_df.head()
```

```python
# merge linking table with compustat data
working_merge = compustat_df.merge(link, how='left',on=['fyear','GVKEY'])
working_merge.head()
```

```python
# read in crsp data
crsp_df = pd.read_csv('quant_analysis/f80d5a3831bae497.csv')
crsp_df['fyear'] = pd.to_datetime(crsp_df['date'], format='%Y%m%d').dt.to_period('Y')
crsp_df['date'] = pd.to_datetime(crsp_df['date'], format='%Y%m%d').dt.to_period('d')
crsp_df
```

```python
# merge crsp data with compustat data
final_merge = working_merge.merge(crsp_df, how='inner',on=['fyear','PERMNO'])
final_merge.shape
```

```python
final_merge.head()
```

```python
# calculate adjusted shares outstanding and adjusted share price
final_merge['SHROUT_ADJ'] = (final_merge['SHROUT'] * final_merge['CFACSHR']) * 1000
final_merge['PRC_ADJ'] = final_merge['PRC'] / final_merge['CFACPR']
```

```python
import copy
# create the data frame to work with
df = final_merge[['date','fyear','tic','PERMNO','COMNAM','bkvlps','PRC_ADJ','SHROUT_ADJ']].copy(deep=True)
df.head()
```

```python
# calculate price-to-book ratio and daily return for each stock
df['pb_ratio']=df['PRC_ADJ']/df['bkvlps']
df['RETURN'] = (df['PRC_ADJ'] / df['PRC_ADJ'].shift(1) - 1)

# drop NaNs and limit P/B to maximum of 3
df=df.dropna()
df1=df[df['pb_ratio']<3]
df1.head()
```

```python
df1.head()
```

```python
def rebalance(df,x):
    df['value'] = ((x/20 * df['RETURN']))+x/20
    return df
    
def reconstitute(df,x):
    df = df.sort_values(by='pb_ratio', ascending=False)[0:20].reset_index(drop=True)
    df = rebalance(df,x)
    return df
```

```python
dates = df['date'].unique()
col = ['date','fyear','tic','PERMNO','PRC_ADJ','pb_ratio']
mf = pd.DataFrame(np.zeros((0,len(col))), columns=col)
mf_work = pd.DataFrame(np.zeros((20,len(col))), columns=col)

# start with 10000
x=10000

# Every day reinvest in firms with the lowest P/B ratios (with P/B ratio capped at 3)
for date in dates:
    
    # get data @ time period
    df2 = df1[df1['date']==date][['date', 'PERMNO', 'tic','PRC_ADJ','pb_ratio','RETURN']] 
    mf_work = reconstitute(df2,x)
    
    # new value of portfolio
    x = mf_work['value'].sum()

    mf = mf.append(mf_work).reset_index(drop=True)
```

```python
hpr = ((x-10000)/10000)*100
annualized = (1+hpr)**(1/2) - 1

print('Money invested:   \t\t $ %.2f' % 10000)
print('Value of portfolio after year 2: $ %.2f ' % x)
print('\nHolding period return: %.2f'% hpr,'%')
print('Annualized HPR: %.2f'% annualized,'%')
```
